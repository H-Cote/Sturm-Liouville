# Ecuación de Chebyshev

### Objetivos:

* Describir los polinomios de Chebyshev
* Usar los los polinomios de Chebyshev para resolver un problema físico.

## Introducción

Dada la ecuación de Sturm–Liouville 
 \begin{align}
{d\over dx}\left[p(x){d\over dx}y(x)\right]-q(x)y(x)+\lambda r(x)y(x)=0
\end{align}

Con las condiciones de frontera:

 \begin{align}
\alpha_1 y(a)+\alpha_2 y'(a)=0\\ 
\beta_1y(b)+\beta_2 y'(b)=0 
\end{align}

donde $a<x<b$ 

Podemos obtener la ecuación de diferencial de Chebyshev: 
\begin{align}
\\ \frac{d}{dx}\left[\sqrt{1-x^2}\frac{dy}{dx}\right]+\frac{n^2}{\sqrt{1-x^2}}y=0
\end{align}

donde:

\begin{align}
p(x)&=\sqrt{1-x^2}&
 q(x)&=0\\
r(x)&=\frac{1}{\sqrt{1-x^2}}&
\lambda&=n^2
\end{align}

con
$-1<x<1$ 

Ahora, vamos a escribir la ecuación anterior en una forma más usual. Para esto escribimos la ecuación de Chebyshev de la siguiente manera:
\begin{align}
\\ \frac{d}{dx}\left[\sqrt{1-x^2}y'\right]+\frac{n^2}{\sqrt{1-x^2}}y=0
\end{align}


recordando la fórmula de la derivada del producto de funciones: 

\begin{align}
\\ f'=u'v+uv'
\end{align}

haciendo $u=\sqrt{1-x^2}$ y $v=y'$ obtenemos:

\begin{align}
\\ \frac{d}{dx}\left[\sqrt{1-x^2}y'\right]=\frac{-x}{\sqrt{1-x^2}}y'+\sqrt{1-x^2}y''
\end{align}

sustituyendo:

\begin{align}
\\ \frac{-x}{\sqrt{1-x^2}}y'+\sqrt{1-x^2}y''+\frac{n^2}{\sqrt{1-x^2}}y=0
\end{align}

multiplicamos por $\sqrt{1-x^2}$

\begin{align}
\\ -xy'+(1-x^2)y''+n^2y=0
\end{align}

ordenando:

\begin{align}
\\(1-x^2)y''-xy'+n^2y=0
\end{align}

esta es la forma usual de escribir la ecuación de Chebyshev.






----

Las soluciones a la ecuación de Chebyshev con el eigenvalor $\lambda=n^2$ con $n=0,1,2,3...$, son mejor conocidas como **polinomios de Chebyshev**, denotados por $T_n(x)$, los cuales forman un conjunto ortogonal en el intervalo $-1<x<1$ con respecto a $r(x)$. 

**Recordatorio**: Los eigenvalores son todos positivos (ver notebook "Ecuación de Sturm-Liouville")

Por lo anterior, podemos escribir la ecuación de Chebyshev de las siguientes formas:

\begin{align}
\\(1-x^2)T_n''-xy'+n^2T_n=0
\end{align}

\begin{align}
\\ \frac{d}{dx}\left[\sqrt{1-x^2}\frac{d}{dx}T_n\right]+\frac{n^2}{\sqrt{1-x^2}}T_n=0
\end{align}

## Soluciones

## Fórmula de recurrencia

### Polinomios de Chebyshev de primer orden

Los polinomios de Chebyshev de primer orden están definidos por la siguiente relación de recurrencia:
 \begin{align}
\\T_0(x) = 1 
\\T_1(x) = x
\\T_{n+1}(x) = 2xT_n(x) - T_{n-1}(x)
\end{align}



## Código

```julia
using PyPlot
```

```julia
function pc1(m,x)
    
    n=m+1
    t=zeros(n)
    
   t[1]=1
   t[2]=x
    
    if n==1
        
        return t[1]
        
    elseif n==2
        
        return t[2]
        
    else
    
        for i=3:n
            t[i]=2*x*t[i-1]-t[i-2]
        end

        return t
    end
    
end
```

```julia
pc1(9,2)
```

```julia
pc1(2,8)
```

```julia
#Conjuntos de datos
x = linspace(-1.5, 1.5, 1000)

y=[]
for i=1:length(x)
    yy= pc1(4,x[i])
    y=push!(y,yy)        
end 

#Gráfica

fig,ax = subplots()
ax[:set_ylim](-2, 2.0)
ax[:set_xlim](-1, 1.0)

title("Polinomios de Chebyshev de primer orden")#Título de la gráfica
xlabel("\$x\$") #Etiqueta para el eje X
ylabel(" \$T_n (x)\$")#Etiqueta para el eje Y
 

axhline(0, color="red") 
grid()

plot(x,y,label="\$T\$")

legend(loc="best")
show()
```

La anterior, es la misma gráfica que se obtiene con la implementación en Matlab: https://www.mathworks.com/help/symbolic/chebyshevt.html

### Polinomios de Chebyshev de segundo orden

Los polinomios de Chebyshov de segundo tipo son definidos mediante la relación de recurrencia:



 \begin{align}
\\ U_0(x) = 1
\\ U_1(x) = 2x 
\\ U_{n+1}(x) = 2xU_n(x) - U_{n-1}(x)
\end{align}


## Código

```julia
function pc2(m,x)
    
    n=m+1
    u=zeros(n)
    
   u[1]=1
   u[2]=2*x
    
    if n==1
        
        return u[1]
        
    elseif n==2
        
        return u[2]
        
    else
    
        for i=3:n
            u[i]=2*x*u[i-1]-u[i-2]
        end

        return u
    end
    
end
```

```julia
x = linspace(-1.5, 1.5, 1000)

y=[]
for i=1:length(x)
    yy= pc2(4,x[i])
    y=push!(y,yy)        
end 

fig,ax = subplots()
ax[:set_ylim](-2, 2.0)
ax[:set_xlim](-1, 1.0)

title("Polinomios de Chebyshev de segundo orden")#Título de la gráfica
xlabel("\$x\$") #Etiqueta para el eje X
ylabel(" \$U_n (x)\$")#Etiqueta para el eje Y

axhline(0, color="red") 
grid()

plot(x,y,label="\$U\$")
legend(loc="best")
show()
```
